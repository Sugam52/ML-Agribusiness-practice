```python
# Grouping by 'size' and summing the 'income' for each group
grouped_data = cake.groupby('size')['income'].sum()

# Bar chart
plt.figure(figsize=(8, 6))
grouped_data.plot(kind='bar', color='skyblue')
plt.title('Income by Size')
plt.xlabel('Size')
plt.ylabel('Income')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# Pie chart
plt.figure(figsize=(8, 8))
grouped_data.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'lightblue'])
plt.title('Income Distribution by Size')
plt.ylabel('')  # Remove y-axis label for pie chart
plt.legend(grouped_data.index, loc='upper right')  # Add legend with 'size' categories
plt.show()
```
