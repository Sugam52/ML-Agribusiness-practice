```python
# generate a category data 1= 1992_1999, 2 = 2000-2009. 3= 2010 after 
# tried by myself didnt worked

gdp_cat =gdp['Year']
gdp_cat.describe
gdp_cat.hist()

gdp['period']=2
gdp['period'][gdp['Year']<2000]="1"
```

```python
# categorising acc to year
import pandas as pd
gdp_df =pd.DataFrame(gdp)
gdp_df['period1']= pd.cut(gdp['Year'],
                          bins=[1980,1999,2009,float('Inf')],
                          labels=['Ancient','Older','Modern'])
gdp_df
```

```python
gdp_df.iloc[:,1:4].mean().plot(kind="bar") # mean of data # 1:4 is column 
                                              # 1:6,0:2 first two columns
                                              
```

```python
# plot by group \ column named period1 
gdp_df.groupby(by='period1').mean().plot(kind="bar")
```

```python
# correlation matrix on plot
country_corr =gdp_df.iloc[:,1:4].corr()
country_corr
```

```python
# now plot corr
import statsmodels.api as sm
sm.graphics.plot_corr(country_corr,xnames=list(country_corr.columns))
plt.show()
sns.heatmap(country_corr,annot=True)
```

```python
## how to run a simple ols
y=gdp['China']
x=gdp[['Germany','USA','Year']]
x=sm.add_constant(x)
ols_gdp=sm.OLS(y,x).fit()
print(ols_gdp.summary())
```

```python
# taking log for gdp
y1=np.log(gdp['China'])
x1=np.log(gdp[['Germany','USA','Year']])
x1=sm.add_constant(x1)
ols_gdp1=sm.OLS(y1,x1).fit()
print(ols_gdp1.summary())
```
